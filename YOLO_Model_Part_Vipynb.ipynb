```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

results = model.train(
    data="/content/visdrone.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    project="/content/runs/detect",
    name="yolov8n_visdrone_fixed",
    exist_ok=True
)
```
